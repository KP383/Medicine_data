```python
from bs4 import BeautifulSoup
```

```python
medicine_image = []
for i in range(1, 987):
    openfile = open(f"/Users/parthkalathia/Desktop/Medicine DATA/All Medicine HTML/A/A_{i}.txt", "r")
    file = openfile.read()
    Parse = BeautifulSoup(file, 'xml')
    medicine = Parse.find_all("p", class_="Card__productName__qw2CE bodyMedium")
    link = Parse.find_all("img")
    
    m=0
    for n in range(0,len(link)):
        if(m == len(medicine)):
            break
        if(medicine[m].text == link[n].attrs['alt']):
            m += 1
            medicine_image.append(link[n].attrs['src'])
    print(len(medicine), len(link))
    print(f"Done file{i}")
```
